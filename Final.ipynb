```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C://Users//USER//Documents//users.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
#  Ages with the highest income
highest_income_ages = df.groupby('Age')['Total Income'].sum().sort_values(ascending=False)
highest_income_ages.plot(kind='bar', title='Ages with the Highest Income')
plt.savefig('highest income_ages.png')
plt.show()
```

```python
# Visualization 2: Gender distribution across spending categories
gender_spending = df.groupby('Gender')[['Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']].sum()
gender_spending.plot(kind='bar', stacked=True, title='Gender Distribution Across Spending Categories')
plt.savefig('gender_spending_distribution.png')
plt.show()
```
